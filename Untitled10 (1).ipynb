```python
import numpy as np
import xarray as xr
import math
from glob import glob
from datetime import datetime, timedelta
import matplotlib as mpl
import matplotlib.pyplot as plt
import cmocean as cm
```

```python
nsofs = 11   #...pick the sofs mission
f = df + "ocean/" + "sofs." + str("%02d" % (nsofs,))
teT = xr.open_mfdataset((f + ".T.nc"))[["temp"]]
tiT = xr.open_mfdataset((f + ".T.nc"))[["time"]]
deT = xr.open_mfdataset((f + ".T.nc"))[["depth"]]
```

```python
teS = xr.open_mfdataset((f + ".S.nc"))[["temp"]]
saS = xr.open_mfdataset((f + ".S.nc"))[["salt"]]
tiS = xr.open_mfdataset((f + ".S.nc"))[["time"]]
deS = xr.open_mfdataset((f + ".S.nc"))[["depth"]]
```

```python
#...i don't know how to deal with xarrays!!
teT = np.asarray(teT['temp'])
teS = np.asarray(teS['temp'])
saS = np.asarray(saS['salt'])
```

```python
#...doesn't convert missing values to NaNs
#   probably because I didn't put a missing value att in the netcdf
teT[teT>100] = np.nan
teS[teS>100] = np.nan
saS[saS>100] = np.nan
```

```python
NtT,NzT = np.shape(teT)
NtS,NzS = np.shape(teS)
```

```python
#...plot stuff
#   note that ive kept the "thermistor string" data separate from
#   the CT data, as they're on different write frequencies
#   surface temp is only in the CT data, so here I just plot
#   the surface temp from the CT sensors over the top
Tmn = 8
Tmx = 15
Smn = 34.5
Smx = 35
tmn = tiS['time'][0]
tmx = tiS['time'][NtS-1]
zmx = 500
Tpal = 'viridis'
Spal = 'Spectral_r'
fig = plt.figure(figsize=(20, 10))
plt.cla()
p=plt.pcolor(tiT['time'],-deT['depth'],teT.T,vmin=Tmn,vmax=Tmx,cmap=Tpal)
plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][0]),NtS),c=teS[:,0],vmin=Tmn,vmax=Tmx,marker='s',s=50,cmap=Tpal)
plt.colorbar(p, extend='both',label='Temp')
for n in np.arange(NzS):
 ii = ~np.isnan(saS[:,n])
 plt.scatter(tiS['time'][ii],np.repeat(np.ravel(-deS['depth'][n]),NtS)[ii],c="grey",marker='s',s=10)
 p=plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c=saS[:,n],vmin=Smn,vmax=Smx,marker='s',s=2,cmap=Spal)
plt.colorbar(p, extend='both',label='Salt',location='left')
ax = plt.gca()
ax.set_xlim([tmn,tmx])
ax.set_ylim([-zmx,10])
```

```python
#...now zoom in on an section of interest
#   when there's a front crossing the mooring is easy to see from the sudden increase in temp and salt
#   at all depths.  occassionally wafting past are some weird sub-thermocline warming events where the signal 
#   arrives first at depth. BUT, in between these times, maybe we can assume that what is going on is basically
#   1D vertical air-sea stuff.
#
#   The section below, from SOFS 11 in 2022, seems like a good example.  From Nov 8 to mid December
#   heat is being pumped downwards, with a couple on interesting cooling events along the way.

Tmn = 9
Tmx = 12
Smn = 34.5
Smx = 34.7
#tmn = tiS['time'][0]
#tmx = tiS['time'][NtS-1]
tmn = datetime(2022,11, 1)
tmx = datetime(2023, 1, 1)
zmx = 200
Spal = 'viridis'
Tpal = 'Spectral_r'
fig = plt.figure(figsize=(20, 10))
plt.cla()
p=plt.pcolor(tiT['time'],-deT['depth'],teT.T,vmin=Tmn,vmax=Tmx,cmap=Tpal)
plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][0]),NtS),c=teS[:,0],vmin=Tmn,vmax=Tmx,marker='s',s=50,cmap=Tpal)
plt.colorbar(p, extend='both',label='Temp')
for n in np.arange(NzS):
 plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c="white",marker='s',s=10)
 p=plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c=saS[:,n],vmin=Smn,vmax=Smx,marker='s',s=5,cmap=Spal)
plt.colorbar(p, extend='both',label='Salt',location='left')
ax = plt.gca()
ax.set_xlim([tmn,tmx])
ax.set_ylim([-zmx,10])
```

```python
#...now zoom in even more on the various heating/cooling episodes
#   over timescales of a few days
#   it's interesting that on the days with surface cooling (15th, 23rd)
#   is when you get warming/deepening at the base of the thermocline 
#   it that convection or wind-driven???

Tmn = 9
Tmx = 10.5
Smn = 34.5
Smx = 34.6
#tmn = tiS['time'][0]
#tmx = tiS['time'][NtS-1]
tmn = datetime(2022,11, 8)
tmx = datetime(2022,12, 8)
zmx = 150
Spal = 'viridis'
Tpal = 'Spectral_r'
fig = plt.figure(figsize=(20, 10))
plt.cla()
p=plt.pcolor(tiT['time'],-deT['depth'],teT.T,vmin=Tmn,vmax=Tmx,cmap=Tpal)
plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][0]),NtS),c=teS[:,0],vmin=Tmn,vmax=Tmx,marker='s',s=80,cmap=Tpal)
plt.colorbar(p, extend='both',label='Temp')
for n in np.arange(NzS):
 plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c="white",marker='s',s=10)
 p=plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c=saS[:,n],vmin=Smn,vmax=Smx,marker='s',s=5,cmap=Spal)
plt.colorbar(p, extend='both',label='Salt',location='left')
ax = plt.gca()
ax.set_xlim([tmn,tmx])
ax.set_ylim([-zmx,1])
```

```python
#...nearby glorys reanalysis data for reference
#   we wouldn't expect the timings of front crossings to match
#   but the structure in between times should
#   the file contains data on a small grid around the mooring location (roughly)
#   here i'm arbitrarily plotting one corner, assuming it doesn't change much
#
#   it actually looks amazingly good.  it's possible i guess that they are assimilating
#   these obs.  but still pretty good!

teg = xr.open_mfdataset(f"/g/data/jk72/mxr581/SOTS-project-data/glorys/cmems*my*.nc")[["thetao"]]
sag = xr.open_mfdataset(f"/g/data/jk72/mxr581/SOTS-project-data/glorys/cmems*my*.nc")[["so"]]

fig = plt.figure(figsize=(20, 10))
plt.cla()
p=plt.pcolor(teg['time'],-teg['depth'],teg['thetao'][:,:,0,0].T,vmin=Tmn,vmax=Tmx,cmap=Tpal)
#plt.scatter(sag['time'],np.repeat(np.ravel(-deS['depth'][0]),NtS),c=teS[:,0],vmin=Tmn,vmax=Tmx,marker='s',s=80,cmap=Tpal)
plt.colorbar(p, extend='both',label='Temp')
#for n in np.arange(NzS):
# plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c="white",marker='s',s=10)
# p=plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c=saS[:,n],vmin=Smn,vmax=Smx,marker='s',s=5,cmap=Spal)
plt.colorbar(p, extend='both',label='just for spacing!',location='left')
ax = plt.gca()
ax.set_xlim([tmn,tmx])
ax.set_ylim([-zmx,1])
```

```python
#...but glorys salinity is off, compared to sofs
#   tends to be 0.1 psu fresher, but more tellingly,
#   doesn't get the salinity changes in synch like
#   it does with temp. a sign that glorys is cheating with temp?!

teg = xr.open_mfdataset(f"/g/data/jk72/mxr581/SOTS-project-data/glorys/cmems*my*.nc")[["thetao"]]
sag = xr.open_mfdataset(f"/g/data/jk72/mxr581/SOTS-project-data/glorys/cmems*my*.nc")[["so"]]

fig = plt.figure(figsize=(20, 10))
plt.cla()
p=plt.pcolor(sag['time'],-sag['depth'],sag['so'][:,:,0,0].T,vmin=34.4,vmax=34.6,cmap=Tpal)
#plt.scatter(sag['time'],np.repeat(np.ravel(-deS['depth'][0]),NtS),c=teS[:,0],vmin=Tmn,vmax=Tmx,marker='s',s=80,cmap=Tpal)
plt.colorbar(p, extend='both',label='Salt glorys')
for n in np.arange(NzS):
 plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c="white",marker='s',s=10)
 p=plt.scatter(tiS['time'],np.repeat(np.ravel(-deS['depth'][n]),NtS),c=saS[:,n],vmin=34.5,vmax=34.6,marker='s',s=5,cmap=Tpal)
plt.colorbar(p, extend='both',label='Salt observed (note different scale)',location='left')
ax = plt.gca()
ax.set_xlim([tmn,tmx])
ax.set_ylim([-zmx,1])
```

```python
#...access om2 iaf run, compared to sofs7
#
#   again, no reason for the model to match the obs, but...
#   it looks like, in this simulation, the sofs site is constantly
#   being flushed by fronts (where temp and salt increase all thru
#   the water column). but in between times, doesn't look at all realistic
#   does glorys only do so well cos it's "cheating"??
#   or is there some major difference between the closure schemes in mom and nemo
#   that could explain it???

nsofs = 7   #...pick the sofs mission
f = df + "ocean/" + "sofs." + str("%02d" % (nsofs,))
sofsT= xr.open_mfdataset((f + ".T.nc"))
temp = np.asarray(sofsT['temp'])
temp[temp>100] = np.nan
sofsS= xr.open_mfdataset((f + ".S.nc"))
salt = np.asarray(sofsS['salt'])
salt[salt>100] = np.nan

tea = xr.open_mfdataset(
    f"/g/data/cj50/access-om2/raw-output/access-om2-01/01deg_jra55v140_iaf/output2[234]*/ocean/ocean-3d*daily*.nc",
    parallel=True, chunks='auto')[["temp"]].sel(
        xt_ocean=141.5-360, yt_ocean=-47, method="nearest")
saa = xr.open_mfdataset(
    f"/g/data/cj50/access-om2/raw-output/access-om2-01/01deg_jra55v140_iaf/output2[234]*/ocean/ocean-3d*daily*.nc",
    parallel=True, chunks='auto')[["salt"]].sel(
        xt_ocean=141.5-360, yt_ocean=-47, method="nearest")
```
